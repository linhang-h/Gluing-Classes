```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from Gluing import GluingSet, GluingLink

alpha = math.sqrt(2)-1
star1 = (alpha/2) % 0.5
star2 = star1 + 0.5
radius = 5
```

```python
def itinerary(x,n):
    resur = ''
    for i in range(n):
        if x > star1 and x < star2:
            resur += 'L'
        elif x == star1 or x == star2:
            resur += '*'
        else:
            resur += 'R'
        x = (x*2) % 1
    return resur

def match(w1, w2):
    if len(w1) > len(w2):
        temp = w1
        w1 = w2
        w2 = temp

    resur = True
    for i in range(len(w1)):
        if w1[i] != '*' and w2[i] != '*':
            if w1[i] != w2[i]:
                resur = False
    return resur

def legalize(w):
    n = len(w)
    lst = list(w)
    Ia = itinerary(alpha, n)
    for i in range(n-2,-1,-1):
        if match(lst[i+1:],Ia):
            lst[i] = '*'
    return ''.join(lst)
```

```python
w = legalize('LLRLLRL')
w
```

```python
links = GluingSet(w)
print(links)
```
